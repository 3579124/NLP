## Exercise 1: Function Plotting and Analytical Determination of Critical Points

Determine _analytically_ the coordinates and nature of the critical points of the following functions:

- $$f_1(x,y) = x^3 + y^3 -3x -12y$$
- $$f_2(x,y) = x^3 -12xy + 8y^3$$
- $$f_3(x,y) = {(x-1)}^2 + 10 {(x^2-y)}^2$$

**_Vocabulary:_** An "analytical" determination means that the calculation is done "by hand", without using numerical tools or formal computation.

```python
# Import numpy and pyplot libraries
import numpy as np
import math 
import matplotlib.pyplot as plt
```

```python
def f1(x,y) :
    output = x ** 3 + y ** 3 - 3 *x -12 *y
    return output
def f2(x,y) :
    output = x ** 3 - 12 * x * y + 8 * (y **3)
    return output
def f3(x,y) :
    output = ( x -1 )** 2 + 10 * (x ** 2 - y) **2
    return output
```

Complete this instruction block so that _df1_ does what is expected. Then use this function to verify that the gradient of $f_1$ is indeed zero at the critical points.

```python
def df1 (x,y) :
# Partial derivative with respect to x
    df_dx = 3 * x ** 2 - 3
# Partial derivative with respect to y 
    df_dy = 3 * y ** 2 - 12
    return [df_dx,df_dy]
```

**3.4** The instruction block below provides the template for the function _d2f1_, which calculates and returns the Hessian matrix of $f_1$. Complete this instruction block so that _d2f1_ does what is expected. Then use this function to verify the nature of the critical points of $f_1$.

```python
# Instruction block to be completed
def d2f1 (x,y) :
    # Second-order partial derivatives
    d2f_dxdx = 6 * x
    d2f_dxdy = 0
    d2f_dydx = 0
    d2f_dydy = 6 * y
    return [[d2f_dxdx,d2f_dxdy],[d2f_dydx,d2f_dydy]]
```

```python
# Template for calculating eigenvalues of a matrix, to be adapted:
matrice = [[1,2],[3,4]]
valProp = np.linalg.eigvals(matrice)
print('Eigenvalues of the matrix {} :'.format(matrice))
print('  vp1 = {}'.format(valProp[0]))
print('  vp2 = {}'.format(valProp[1]))
```

```python
# Determine the nature of the critical points of f1 using the Hessian
crit_points = [(-1, -2), (-1, 2), (1, -2), (1, 2)]
for pt in crit_points:
    H = np.array(d2f1(*pt), dtype=float)
    eig = np.linalg.eigvals(H)
    # Classification: all positive -> min, all negative -> max, mixed -> saddle
    if np.all(eig > 0):
        nature = 'Local minimum'
    elif np.all(eig < 0):
        nature = 'Local maximum'
    else:
        nature = 'Saddle point'
    print(f"Point {pt}: Hessian = {H.tolist()}, eigenvalues = {np.sort(eig).tolist()} -> {nature}")
```

**3.5 - Gradient function for f2**

Define the gradient function for $f_2$ and verify it's zero at critical points.

```python
def df2(x,y):
    # Partial derivative with respect to x: d/dx(x^3 - 12xy + 8y^3) = 3x^2 - 12y
    df_dx = 3 * x**2 - 12 * y
    # Partial derivative with respect to y: d/dy(x^3 - 12xy + 8y^3) = -12x + 24y^2
    df_dy = -12 * x + 24 * y**2
    return [df_dx, df_dy]
```

```python
# Critical points for f2: (0,0) and (2,1)
crit_points_f2 = [(0, 0), (2, 1)]
print("Gradient verification at critical points of f2:")
for pt in crit_points_f2:
    g = df2(*pt)
    print(f"df2{pt} = {g}")
```

**3.5 - Hessian function for f2**

Define the Hessian matrix function for $f_2$ and determine the nature of critical points.

```python
def d2f2(x,y):
    # df/dx = 6x
    d2f_dxdx = 6 * x
    # df/dxdy = -12
    d2f_dxdy = -12
    # df/dydx = -12 (symmetry)
    d2f_dydx = -12
    # df/dy = 48y
    d2f_dydy = 48 * y
    return [[d2f_dxdx, d2f_dxdy], [d2f_dydx, d2f_dydy]]
```

```python
# Determine the nature of the critical points of f2 using the Hessian
crit_points_f2 = [(0, 0), (2, 1)]
print("Nature of critical points for f2:\n")
for pt in crit_points_f2:
    H = np.array(d2f2(*pt), dtype=float)
    eig = np.linalg.eigvals(H)

    if np.all(eig > 0):
        nature = 'Local minimum'
    elif np.all(eig < 0):
        nature = 'Local maximum'
    else:
        nature = 'Saddle point'
    print(f"Point {pt}:")
    print(f"  Hessian = {H.tolist()}")
    print(f"  Eigenvalues = {np.sort(eig).tolist()}")
    print(f"  Nature: {nature}\n")
```
