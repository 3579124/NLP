### ðŸ“š Question 1: Understanding Pipelines
Before we start using the models, let's understand what we're working with:

1. What is a pipeline in Hugging Face Transformers? What does it abstract away from the user?  
  
    -> A pipeline in Hugging Face Transformers is a high-level API that provides an end-to-end interface for common NLP (and multimodal) tasks. It abstracts away the complexity of the full inference workflow by automatically handling: Input preprocessing (e.g. tokenization, normalization, truncation); Model loading and inference; Output post-processing (e.g. decoding, probability or label formatting).As a result, users can perform inference with minimal code without needing to manage low-level details of model architectures or data preparation.  
      
2. Visit the pipeline documentation and list at least 3 other tasks (besides text-classification) that are available.  
  
    -> In addition to text classification, Hugging Face pipelines support a wide range of tasks, including for example: Text generation, Named entity recognition (token classification),Question answering, Summarization, Translation, and Visual question answering. These pipelines cover both text-only and multimodal use cases.  
  
3. What happens when you don't specify a model in the pipeline? How can you specify a specific model?  
  
    -> When no model is explicitly specified, the pipeline automatically loads a default pretrained model that is considered suitable for the chosen task, based on Hugging Faceâ€™s model hub recommendations. To specify a particular model, the user can pass the model argument when creating the pipeline, either by providing a model identifier from the Hugging Face Hub or a local model path. This allows full control over the architecture, training data, and performance characteristics used during inference.

### ðŸ“š Question 3: Named Entity Recognition (NER)
Let's understand NER better:

1. What does the aggregation_strategy="simple" parameter do in the NER pipeline? Check the token classification documentation.  
  
    -> The aggregation_strategy="simple" parameter groups consecutive tokens that share the same entity label into a single entity span. Instead of returning predictions at the subword or token level, it merges them to produce more human-readable entity outputs corresponding to complete words or phrases.  
      
2. Looking at the output above, what do the entity types mean? (ORG, MISC, LOC, PER)  
  
    -> These entity types correspond to standard Named Entity Recognition categories:  
    ORG: Organization (e.g. companies, institutions)  
    MISC: Miscellaneous entities that do not fit other predefined categories  
    LOC: Location (e.g. countries, cities, geographical areas)  
    PER: Person (individual human names)  

3. Why do some words appear with ## prefix (like ##tron and ##icons)? What does this indicate about tokenization?  
  
    -> The ## prefix indicates subword tokens produced by WordPiece tokenization. It means that the token is a continuation of the previous token rather than the beginning of a new word. This approach allows the tokenizer to handle rare or unseen words by breaking them into smaller, known subword units.  
      
4. The model seems to have split "Megatron" and "Decepticons" incorrectly. Why might this happen? What does this tell you about the model's training data?  
  
    -> This happens because the model was trained primarily on real-world entities found in the CoNLL-2003 dataset, which consists of news articles. Fictional names such as â€œMegatronâ€ and â€œDecepticonsâ€ are unlikely to appear in the training data. As a result, the tokenizer splits them into subword units, and the model may struggle to recognize them as coherent named entities. This highlights the dependence of NER performance on the domain and coverage of the training corpus.  
  
5. Challenge: Find the model card for dbmdz/bert-large-cased-finetuned-conll03-english. What is the CoNLL-2003 dataset?Question AnsweringÂ¶  
  
    -> The CoNLL-2003 dataset is a widely used benchmark dataset for Named Entity Recognition. It consists of English news articles annotated at the token level with four entity types: PER, LOC, ORG, and MISC. The dataset was introduced as part of the CoNLL-2003 shared task and remains a standard evaluation benchmark for NER models.  
      
ðŸ¤” How might the choice of tokenizer affect NER performance?  
  
  The choice of tokenizer can significantly impact NER performance because it determines how words are segmented into tokens. Subword-based tokenizers (such as WordPiece or BPE) improve robustness to unseen words but can fragment entity names, making correct entity boundary detection more difficult. Conversely, tokenizers that better preserve whole words may improve entity span recognition but at the cost of reduced flexibility for rare or unknown terms. Therefore, tokenizer design directly affects both precision and recall in NER tasks.

### ðŸ“š Question 4: Question Answering Systems
Explore how question answering works:

1. What type of question answering is this? (Extractive vs. Generative) Check the question answering documentation.  
  
    -> This is extractive question answering. In extractive QA, the model selects an exact text span from the provided context as the answer, rather than generating new text. The answer must already appear verbatim in the input passage.  
      
2. The model outputs start and end indices. What do these represent? Why are they important?  
   
    -> The start and end indices represent the positions of the answer span within the input context. They indicate where the predicted answer begins and ends, typically at the token or character level. These indices are crucial because extractive QA models do not generate answers; instead, they identify and extract the most likely span corresponding to the question.  
      
3. What is the SQuAD dataset? (Look up the model distilbert-base-cased-distilled-squad on the Hub)  
      
    -> The Stanford Question Answering Dataset (SQuAD) is a large-scale benchmark dataset for extractive QA. It contains over 100,000 questionâ€“answer pairs based on Wikipedia articles, where each answer corresponds to a contiguous span of text within the given context. Models such as distilbert-base-cased-distilled-squad are fine-tuned on this dataset.  
      
4. Try to think of a question this model CANNOT answer based on the text. Why would it fail?ðŸ’¡ Try asking questions that require reasoning or information not in the text. What happens?  
  
    -> An example of a question the model cannot answer is: â€œWhat should Amazon do?â€ This question requires reasoning, interpretation, or opinion, rather than factual extraction. Since extractive QA models are limited to selecting spans that explicitly appear in the input text, they fail when the answer is not directly stated. In such cases, the model may either return an unrelated span or produce an incoherent result. This limitation highlights the difference between extractive and generative QA approaches.

Before running the summarization code, let's understand how it works:

1. What is the difference between extractive and abstractive summarization? Check the summarization documentation.  
   
    -> Extractive summarization selects and concatenates existing sentences or phrases directly from the input text to form a summary. Abstractive summarization, in contrast, generates new text that paraphrases and condenses the original content, potentially using words and sentence structures that do not explicitly appear in the source.

3. What do the max_length and min_length parameters control? What happens if min_length > max_length?  
  
    -> The max_length and min_length parameters control the maximum and minimum number of tokens allowed in the generated summary. If min_length is set higher than max_length, the generation process fails or raises an error because the length constraints are contradictory and cannot be satisfied simultaneously.  
      
4. The parameter clean_up_tokenization_spaces=True is used. What does this parameter do? Why might it be useful for summarization?  
  
    -> The clean_up_tokenization_spaces=True parameter removes unnecessary whitespace introduced during tokenization and decoding, such as extra spaces before punctuation. This improves the readability and grammatical cleanliness of the generated summary, which is especially important for user-facing summarization outputs.

ðŸ’¡ Why might summarization be more challenging than text classification? What linguistic capabilities does the model need?  
Summarization is more challenging than text classification because it requires deep semantic understanding, content selection, and natural language generation. The model must identify the most salient information, maintain factual consistency, preserve coherence, and generate fluent, concise text. In contrast, text classification only requires mapping an input to a fixed set of labels without generating new language.

```python
summarizer = pipeline("summarization")
outputs = summarizer(text, max_length=45, clean_up_tokenization_spaces=True)
print(outputs[0]['summary_text'])
```
